Menghitung missing value dan presentase nya

```python
missing_table = pd.DataFrame({
    'Missing value': df.isnull().sum(),
    'Missing (%)': (df.isnull().mean() * 100).round(2)
})

missing_table.sort_values('Missing (%)', ascending=False)
```

Data Cleaning

1. Menghapus Kolom dengan Missing Value >70%

```python
# Ambil list kolom yang missing > 70%
cols_to_drop = missing_table[missing_table['Missing (%)'] > 70].index

print("Kolom yang dihapus:", list(cols_to_drop))

# Hapus kolom tersebut dari dataframe
df = df.drop(columns=cols_to_drop)
```

2. Menghapus Baris dengan Missing Value pada kolom Price

```python
# Cek berapa banyak baris yang Price-nya kosong
df['Price'].isna().sum()

# Drop baris yang Price-nya NaN
df = df.dropna(subset=['Price'])
```

```python
# cek sudah tidak ada missing value di kolom Price
df['Price'].isna().sum()
```

3. Mengisi Missing Value
4. Mengubah tipe data

```python
df.info()
```

```python
# Daftar kolom berisi jumlah bilangan bulat yang ingin diubah ke integer
int_cols = [
    'TotalRooms',
    'NumberOfBedrooms',
    'NumberOfWC',
    'NumberOfBathrooms',
    'Parking',
    'ConstructionYear'
]

# Isi missing value dengan median
for col in int_cols:
    df[col] = df[col].fillna(df[col].median())

# Convert float ke int
for col in int_cols:
    df[col] = df[col].astype(int)

# Cek hasil
df[int_cols].dtypes
```

```python
# Kolom TotalArea dan LivingArea diimputasi menggunakan median karena keduanya adalah data numerik yang berisi ukuran area,
# maka Median lebih stabil dibanding mean untuk digunakan.
df['TotalArea'] = df['TotalArea'].fillna(df['TotalArea'].median())
df['LivingArea'] = df['LivingArea'].fillna(df['LivingArea'].median())
```

```python
# imputasi kolom bertipe object dengan mode
object_cols = df.select_dtypes(include='object').columns

for col in object_cols:
    if df[col].isna().sum() > 0:   # hanya imputasi jika ada missing
        df[col] = df[col].fillna(df[col].mode()[0])
```

```python
df.info()
```

5. Menyimpan Dataset Final setelah melakukan cleaning data

```python
# Simpan ke CSV
df.to_csv('portugal_cleaned.csv', index=False)
```
